```python
#e importan las bibliotecas necesarias: tensorflow y numpy.
import tensorflow as tf
import numpy as np
```

```python
#Se definen dos arrays NumPy, celsius y farenheit
celsius=np.array([-40,-25,-10,0,10,25,35,50])
farenheit=np.array([-40,-13,14,32,50,77,95,122])
```

```python
#CON UNA NEURONA
capa=tf.keras.layers.Dense(units=1,input_shape=[1])#Se crea una capa de red neuronal densa (Dense) con una unidad y una forma de entrada de tamaño 1
modelo=tf.keras.Sequential([capa])#Se crea un modelo secuencial (Sequential) de TensorFlow y se agrega la capa anterior al modelo.
```

```python
modelo.compile(#Se compila el modelo utilizando el optimizador Adam con una tasa de aprendizaje de 0.1 y se utiliza la función de pérdida de error cuadrático medio
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
historia=modelo.fit(celsius,farenheit,epochs=600)
print("Modelo entrenado")
```

```python
# Realizar la predicción
fahrenheit_value = modelo.predict([30,20,15])

# Imprimir el resultado
print(fahrenheit_value)
```
